```python
import fitsio
from astropy.io import fits
from astropy.table import Table
import numpy as np
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plot
%matplotlib inline
```

```python
cosmos_path = "/global/project/projectdirs/des/y3-image-sims/input_cosmos_v4.fits"
cosmos_table=Table(fits.open(cosmos_path)[1].data.newbyteorder().byteswap())
```

```python
gold_df = pd.DataFrame()

n_gold = 43
gold_path_list = [f"/global/cscratch1/sd/myamamot/gold/gold_highconf_galaxies_{i+1:06d}.fits" for i in range(n_gold)]
for gold_path in tqdm(gold_path_list):
    gold_table=Table(fits.open(gold_path)[1].data.newbyteorder().byteswap())
    gold_df = pd.concat([gold_df, gold_table.to_pandas()])
```

```python
gold_path = "/global/cscratch1/sd/myamamot/gold/gold_highconf_galaxies_000001.fits"
gold_table=Table(fits.open(gold_path)[1].data.newbyteorder().byteswap())
```

```python
laigle_path = "/global/project/projectdirs/des/zhou/image_simulation/COSMOS2015_Laigle+_v1.1.fits"
laigle_fits = fitsio.FITS(laigle_path)
```

```python
laigle_recarray = laigle_fits[1].read(columns=['NUMBER','ALPHA_J2000','DELTA_J2000'])

laigle_df = pd.DataFrame(laigle_recarray.newbyteorder().byteswap())

laigle_df.set_index("NUMBER",inplace=True)
```

```python
cosmos = cosmos_table.to_pandas()

cosmos.set_index('laigle_number',inplace=True)

cosmos_laigle = pd.merge(cosmos, laigle_df, left_index=True, right_index = True, how="inner")

assert print(cosmos_laigle.loc[219743] ['ALPHA_J2000']) == print(laigle_df.loc[219743]['ALPHA_J2000'])
```
